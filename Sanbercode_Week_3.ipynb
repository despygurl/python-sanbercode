# **1. PENGENALAN MATPLOTLIB**


---
1. **Figure** = window atau page atau halaman dalam objek visual. kalau kita ngegambar di kertas, maka kertas tersebutlah yang di namakan figure

2. **Axis** = suatu area di dalam figure dimana data akan diplot. Dalam Axis juga ada banyak komponen lain seperti x-axis, y-axis, tick labels dll..

3. **Artist** = komponen yang digunakan untuk memperindah objek visualisasi kita, seperti legend, title, tick labels dan lainnya.

```
import random
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(0, 10, 100)
y = [random.random()*100 for i in range (100)]

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.show()
```

```
# Membuat figure
import matplotlib.pyplot as plt

fig = plt.figure()
print(fig)
```

```
# LANGKAH MEMBUAT TAMPILAN AWAL

# Membuat objek figure
fig = plt.figure()

# Membuat objek axes
ax = fig.add_subplot()

# Menampilkan objek visual
plt.show()
```

```
# MEMBUAT PLOT SEDERHANA (Cara 1)

# Membuat objek figure
fig = plt.figure()

# Membuat objek axes
ax = fig.add_subplot()

# Membuat data
data_x = [1,2,3,4,5]
data_y = [10,20,25,30,15]

# Menambahkan ke dalam objek axes
ax.plot(data_x, data_y)

# Menampilkan data
plt.show()
```

```
# MEMBUAT PLOT SEDERHANA (Cara 2)

# Membuat objek visual
plt.plot(data_x, data_y)

# Menampilkan objek visual
plt.show()
```

```
# MENYINGKAT LANGKAH PEMBUATAN PLOT DENGAN MENDEFINISIKAN FIGURE & AXES DALAM SEKALI WAKTU

# Membuat objek figure dan axes
fig, ax = plt.subplots()

# Membuat data
data_x = [1, 2, 3, 4, 5]
data_y = [10, 20, 25, 30, 15]

# Menambahkan dalam objek axes
ax.plot(data_x, data_y)

# Menampilkan data
plt.show()
```

```
# MEMPLOT LEBIH DARI 1 DATA

fig, ax = plt.subplots()

data_x1 = [1,2,3,4]
data_y1 = [10,20,25,30]

data_x2 = [1,2,3,4,5]
data_y2 = [5,15,20,25,30]

ax.plot(data_x1, data_y1)
ax.plot(data_x2, data_y2)

plt.show()
```

```
# Daftar Style Plot pada Matplotlib
plt.style.available
```

```
# MENGUBAH STYLE PLOT

fig, ax = plt.subplots()

data_x1 = [1,2,3,4]
data_y1 = [10,20,25,30]

data_x2 = [1,2,3,4,5]
data_y2 = [5,15,20,25,30]

ax.plot(data_x1, data_y1)
ax.plot(data_x2, data_y2)

# Mengubah style plot
plt.style.use('seaborn-notebook')
plt.show()
```

```
# MENYIMPAN HASIL PLOTTING

fig, ax = plt.subplots()

x1 = [1,2,3,4,5]
y1 = [10,20,25,30,35]

x2 = [1,2,3,4,5]
y2 = [5,15,20,25,30]

ax.plot(x1,y1)
ax.plot(x2,y2)

plt.show()

# Menyimpan hasil figure
fig.savefig('coba.png')

# Menampilkan figure yang disimpan
from IPython.display import Image
Image('coba.png')
```

# **2. AKSES/INDEXING DAN TRANSFORMASI DATA**


---



**Indexing Series**

Objek series dapat dilakukan indexing, tapi objek series tersebut harus ditempatkan sebagai dictionary dan array.

***a. Series sebagai Dictionary***

---



```
import pandas as pd

# Membuat object series
obj_series = pd.Series([1,2,3,4], index=['a','b','c','d'])
obj_series
```

```
# Indexing key
obj_series.keys()
```

```
# Indexing value
obj_series.values
```

```
# Checking existing keys
'b' in obj_series
```

***b. Series sebagai Array***

---



```
# Slicing dengan explicit index
obj_series['a':'c']
```

```
# Slicing dengan implicit index
obj_series[:3]
```

***c. Indexing menggunakan .loc dan .iloc objek series***

---



```
# .iloc indexing
obj_series.loc['a':'c']
```

```
# .iloc indexing
obj_series.iloc[:3] 
```

***d. Filtering objek series***

---



```
# Ditentukan dengan komparasi
obj_series[obj_series > 2]
```

```
# Ditentukan langsung dengan menentukan index
obj_series[['c','d']]
```
